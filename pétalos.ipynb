```python
iris['species'].replace(['Iris-setosa','Iris-virginica', 'Iris-versicolor ', 'Iris-setosa  '], [0,1,2,3], inplace=True)
iris
```

```python
iris['sepal_length'] = iris['sepal_length']*(10)
iris['sepal_width'] = iris['sepal_width']*(10)
iris['petal_length'] = iris['petal_length']*(10)
iris['petal_width'] = iris['petal_width']*(10)
```

```python
iris_t = iris.copy()
iris_t
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
data = iris_t.drop('species', axis=1)
target = iris_t['species']
```

```python
data_test, data_train, target_test, target_train = train_test_split(data, target, test_size=0.75, random_state=1)
data_train.shape, data_test.shape, target_train.shape, target_test.shape 
```

```python
clf = KNeighborsClassifier(n_neighbors = 13)
clf.fit(data_train, target_train)
```

```python
scoring = 'accuracy'
score = cross_val_score(clf, data_train, target_train, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
pred = clf.predict(data_test)
```
